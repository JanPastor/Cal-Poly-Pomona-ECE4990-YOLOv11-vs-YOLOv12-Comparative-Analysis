```python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Colab single-cell: Persistent training & resume for all YOLO12 variants
# on VisDrone dataset â€” early stopping disabled (300 epochs)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

# 0. Mount Drive for persistence
from google.colab import drive
drive.mount('/content/drive')

# 1. Install Ultralytics
!pip install ultralytics --upgrade -q

import os
from ultralytics import YOLO

# 2. Base directory on Drive
BASE_DIR = '/content/drive/MyDrive/YOLO12_VisDrone'

# 3. Model variants
variants = ["yolo12n.pt", "yolo12s.pt", "yolo12m.pt", "yolo12l.pt", "yolo12x.pt"]

# 4. Loop through each variant
for variant in variants:
    model_name = variant.replace('.pt', '')
    print(f'\n=== {model_name} ===')

    # Path to checkpoint
    ckpt = os.path.join(BASE_DIR, 'train', model_name, 'weights', 'last.pt')

    # Resume or fresh start
    if os.path.exists(ckpt):
        print(f'Resuming from checkpoint: {ckpt}')
        model = YOLO(ckpt)
        resume_flag = True
    else:
        model = YOLO(variant)
        resume_flag = False

    # 5. Train for exactly 300 epochs (early stopping disabled)
    try:
        model.train(
            data='VisDrone.yaml',
            epochs=300,                  # exactly 300 epochs
            patience=0,                  # disables early stopping
            imgsz=640,
            batch=16,
            resume=resume_flag,
            project=os.path.join(BASE_DIR, 'train'),
            name=model_name,
            exist_ok=True
        )
    except AssertionError:
        print(f"[SKIP] {model_name} already completed training.")

    # 6. Run validation on both val and test splits
    for split in ['val', 'test']:
        try:
            model.val(
                data='VisDrone.yaml',
                split=split,
                imgsz=640,
                batch=16,
                project=os.path.join(BASE_DIR, split),
                name=model_name,
                exist_ok=True
            )
        except Exception as e:
            print(f"[SKIP VAL] {model_name} {split} failed: {e}")
```

```python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Cell 2: Generate test-split CSVs & results.png for all YOLO12 variants
# Includes F1, speed, per-class APs, plots, and fallback handling
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

# 0. Mount Drive for persistence
from google.colab import drive
drive.mount('/content/drive')

# 1. Install Ultralytics
!pip install ultralytics --upgrade -q

import os
from ultralytics import YOLO

# 2. Base directory on Drive
BASE_DIR = '/content/drive/MyDrive/YOLO12_VisDrone'

import os
import pandas as pd
from ultralytics import YOLO
from ultralytics.utils.plotting import plot_results

# Match your training BASE_DIR for YOLOv12
BASE_DIR = '/content/drive/MyDrive/YOLO12_VisDrone'
variants = ["yolo12n.pt", "yolo12s.pt", "yolo12m.pt", "yolo12l.pt", "yolo12x.pt"]

for variant in variants:
    name = variant.replace('.pt', '')
    print(f"\n>>> Processing {name}")

    # 1) Locate weights folder
    wdir = os.path.join(BASE_DIR, 'train', name, 'weights')
    if not os.path.isdir(wdir):
        print(f"  [SKIP] Weights dir missing: {wdir}")
        continue
    print("  Files:", os.listdir(wdir))

    # 2) Choose checkpoint
    best = os.path.join(wdir, 'best.pt')
    last = os.path.join(wdir, 'last.pt')
    if os.path.exists(best):
        ckpt = best
        print(f"  Using best.pt â†’ {ckpt}")
    elif os.path.exists(last):
        ckpt = last
        print(f"  [WARN] best.pt missing, using last.pt â†’ {ckpt}")
    else:
        print(f"  [SKIP] No checkpoint for {name}")
        continue

    # 3) Run test-split evaluation (confusion matrix + PR curve plots enabled)
    stats = YOLO(ckpt).val(
        data='VisDrone.yaml',
        split='test',
        imgsz=640,
        batch=16,
        save=False,
        plots=True
    )

    # 4) Extract metrics
    prec   = stats.box.mp
    rec    = stats.box.mr
    mAP50  = stats.box.map50
    mAP095 = stats.box.map
    f1     = 2 * prec * rec / (prec + rec) if (prec + rec) > 0 else 0
    speed  = stats.speed['inference']
    maps   = stats.box.maps

    # 5) Build row with dummy losses
    row = {
        'epoch':         0,
        'box_loss':      0.0,
        'cls_loss':      0.0,
        'obj_loss':      0.0,
        'precision':     prec,
        'recall':        rec,
        'mAP50':         mAP50,
        'mAP50-95':      mAP095,
        'F1':            f1,
        'infer_time_ms': speed
    }
    for i, cname in enumerate(stats.names.values()):
        row[f'AP50-95_{cname}'] = maps[i]

    df = pd.DataFrame([row])

    # 6) Save CSV
    test_dir = os.path.join(BASE_DIR, 'test', name)
    os.makedirs(test_dir, exist_ok=True)
    csv_path = os.path.join(test_dir, 'results.csv')
    df.to_csv(csv_path, index=False)
    print(f"  â€¢ CSV â†’ {csv_path}")

    # 7) Generate results.png
    plot_results(file=csv_path, dir=test_dir)
    print(f"  â€¢ Plot â†’ {os.path.join(test_dir, 'results.png')}")
```

```python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# Cell 3: Aggregate all per-model YOLOv12 test metrics into a single Excel workbook
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

import os
import pandas as pd

# MUST match your BASE_DIR from Cells 1 & 2
BASE_DIR = '/content/drive/MyDrive/YOLO12_VisDrone'
variants = ["yolo12n", "yolo12s", "yolo12m", "yolo12l", "yolo12x"]

# Collect each model's CSV into one list
all_dfs = []
for name in variants:
    csv_path = os.path.join(BASE_DIR, 'test', name, 'results.csv')
    if not os.path.isfile(csv_path):
        print(f"[WARN] Missing {csv_path}, skipping")
        continue
    df = pd.read_csv(csv_path)
    df.insert(0, 'model', name)  # Add a model identifier column
    all_dfs.append(df)

# Concatenate into one summary DataFrame
summary_df = pd.concat(all_dfs, ignore_index=True)

# Create summary folder
summary_dir = os.path.join(BASE_DIR, 'summary')
os.makedirs(summary_dir, exist_ok=True)

# Write to Excel
excel_path = os.path.join(summary_dir, 'test_metrics_summary.xlsx')
summary_df.to_excel(excel_path, index=False)

print(f"âœ… Combined summary saved to: {excel_path}")
```
